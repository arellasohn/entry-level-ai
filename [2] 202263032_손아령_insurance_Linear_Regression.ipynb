## ※ Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# ▶ Linear regression 구현하기
### 1. insurance.csv 로드
### 2. age, bmi, children을 X로 charges를 y로 할 것
### 3. training data와 test data는 7:3으로 분리할 것
### 4. age, bmi, children 값을 입력받으면 charges를 예측하는 model을 학습할 것
### 5. test data 입력에 대한 예측값을 구할 것

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
ins = df.values
```

```python
X = ins[:, :3]
y = ins[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
print(pred)
```
