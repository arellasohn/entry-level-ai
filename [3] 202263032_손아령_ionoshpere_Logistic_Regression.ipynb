## ※ Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

## 1.ionosphere.csv 로드

```python
df = pd.read_csv('ionosphere.csv')
df.head()
```

```python
ion = df.values
```

```python
# y는 마지막 열, X는 y를 제외한 모든 열
X = ion[:, :32]
y = ion[:, -1]
```

```python
# X shape 출력
print(X.shape)
```

```python
# y shape 출력
print(y.shape)
```

## 2.학습데이터와 테스트 데이터 분리

```python
# 전체 데이터의 80%는 학습 데이터, 20%는 테스트 데이터로 분리
# X_train: 학습데이터의 X
# y_train: 학습데이터의 y
# X_test: 테스트데이터의 X
# y_test: 테스트데이터의 y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# X_train (학습데이터의 X) shape
print(X_train.shape)
```

```python
# y_train (학습데이터의 y) shape
print(y_train.shape)
```

```python
# X_test (테스트데이터의 X) shape
print(X_test.shape)
```

```python
# y_test (테스트데이터의 X) shape
print(y_test.shape)
```

## 3.학습데이터를 이용하여 Logistic Regression model 학습

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

## 4.테스트데이터의 X를 model에 적용하여 예측값(y) 구하기

```python
pred = lr.predict(X_test)
print(pred)
```

## 5.성능 평가

### 5-1. Confusion matrix 구하기

```python
# Confusion matrix 출력
cm = confusion_matrix(y_test, pred)
print(cm)
```

### 5-2. Accuracy와 error rate 구하기

```python
TP = cm[1, 1]
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
```

```python
# Accuracy 출력 (%)
acc = (TP+TN) / (TP+TN+FP+FN) * 100
print('Accuracy:', round(acc, 2), '%')
```

```python
# Error rate 출력 (%)
err = (FP+FN) / (TP+TN+FP+FN) * 100
print('Error rate:', round(err, 2), '%')
```

### 5-3. Precision과 Recall 구하기

```python
# Precision 출력 (%)
precision = TP / (TP+FP) * 100
print('Precision:', round(precision, 2), '%')
```

```python
# Recall 출력 (%)
recall = TP / (TP+FN) * 100
print('Recall:', round(recall, 2), '%')
```
